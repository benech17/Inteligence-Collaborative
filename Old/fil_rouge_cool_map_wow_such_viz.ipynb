```python
import pandas as pd
import folium
```

```python
depots = pd.read_excel('/content/4_detail_table_depots.xls').drop_duplicates(subset = ['DEPOT_CODE']).reset_index()[['DEPOT_NUMBER','DEPOT_CODE','DEPOT_LATITUDE','DEPOT_LONGITUDE']]
depots
```

```python
clients = pd.read_excel('/content/2_detail_table_customers.xls').drop_duplicates(subset = ['CUSTOMER_CODE']).reset_index()[['CUSTOMER_NUMBER','CUSTOMER_CODE','CUSTOMER_LATITUDE','CUSTOMER_LONGITUDE','NUMBER_OF_ARTICLES','TOTAL_WEIGHT_KG','TOTAL_VOLUME_M3']]
clients
```

```python
m = folium.Map(location=depots.iloc[0][['DEPOT_LATITUDE','DEPOT_LONGITUDE']], zoom_start=12)
for index, row in depots.iterrows():
  folium.Marker(row[['DEPOT_LATITUDE','DEPOT_LONGITUDE']], icon=folium.Icon(color="red")).add_to(m)
for index, row in clients.iterrows():
  folium.Marker(row[['CUSTOMER_LATITUDE','CUSTOMER_LONGITUDE']]).add_to(m)
m
```
